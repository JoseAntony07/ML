```python
# Q) Sample problem of predicting home price in monroe, new jersey (USA)?
```

```python
"""
- Below is the table containing home prices in monroe twp, NJ. Here price depends on area (square feet), 
bed rooms and age of the home (in years). Given these prices we have to predict prices of new homes based on area, 
bed rooms and age.
"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn import linear_model
```

```python
image_path = 'homeprices_multivariate.jpg'

img = mpimg.imread(image_path)
plt.imshow(img)
plt.axis('off')  # Hide axes
plt.show()
```

```python
"""
Given these home prices find out price of a home that has,

Q) 3000 sqr ft area, 3 bedrooms, 40 year old?

Q) 2500 sqr ft area, 4 bedrooms, 5 year old?

- We will use regression with multiple variables here. Price can be calculated using following equation,
- Here area, bedrooms, age are called independant variables or features whereas price is a dependant variable

"""
```

```python
image_path = 'equation_multivariate.jpg'

img = mpimg.imread(image_path)
plt.imshow(img)
plt.axis('off')  # Hide axes
plt.show()
```

```python
df = pd.read_csv("homeprices_multivariate.csv")
df
```

```python
# Step we are going to follow

"""

Data Preprocessing - Fill NA values with median value of a column

Linear regression using multiple variables
  
"""
```

```python
df.bedrooms.median()  # fill the value in missed place
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
# we are done the data preprocessing
# next step - train the model
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price)  # fit - training model using our data  x_train, 
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000, 3, 40]])  # Find price of home with 3000 sqr ft area, 3 bedrooms, 40 year old
```

```python
112.06244194*3000 + 23388.88007794*3 + -3231.71790863*40 + 221323.00186540396
```

```python
"""

Key Steps in Data Preprocessing:

- Data Cleaning: Removing or fixing errors, missing values, and inconsistencies in the data.
- Data Transformation: Scaling, normalizing, or encoding data into a suitable format.
- Feature Selection: Choosing the most relevant features (columns) for the analysis.
- Data Splitting: Dividing the data into training and testing sets.

"""
```

```python
%matplotlib inline

# area vs price

plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color="red", marker="+")
plt.plot(df.area, reg.predict(df[['area', 'bedrooms', 'age']]), color='blue')
```

```python
%matplotlib inline

# bedrooms vs price

plt.xlabel('bedrooms', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.bedrooms, df.price, color="red", marker="+")
plt.plot(df.bedrooms, reg.predict(df[['area', 'bedrooms', 'age']]), color='blue')
```

```python
%matplotlib inline

# age vs price

plt.xlabel('age', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.age, df.price, color="red", marker="+")
plt.plot(df.age, reg.predict(df[['area', 'bedrooms', 'age']]), color='blue')
```
