```python
import pandas as pd

df = pd.read_csv('titanic.csv')
df.head()
```

```python
# data preprocessing, remove unrequired features

input_data = df.drop(['Survived', 'PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
input_data
```

```python
target = df['Survived']
target  # output
```

```python
# convert categories to numbers input_data(Ml model only know the numbers)

from sklearn.preprocessing import LabelEncoder

le_sex = LabelEncoder()
```

```python
input_data['sex_n'] = le_sex.fit_transform(input_data['Sex'])
input_data
```

```python
input_data_new = input_data.drop('Sex', axis='columns')
input_data_new
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(input_data_new, target, test_size=0.2)
```

```python
len(X_train), len(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier


model = DecisionTreeClassifier()

model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```

```python
model.predict([[1, 38, 71.2833, 0]])  # Pclass, Age, Fare, sex_n
```

```python
model.predict([[1, 40, 60.2833, 1]])
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt


# Plot the Decision Tree
plt.figure(figsize=(46,40))
plot_tree(
    model, 
    feature_names=input_data_new.columns, 
    class_names=['Not Survived', 'Survived'], 
    filled=True, 
    rounded=True, 
    fontsize=12,  # Increase font size
    proportion=True  # Normalize the branch size
)
plt.show()
```
