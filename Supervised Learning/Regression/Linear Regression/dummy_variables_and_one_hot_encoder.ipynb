<h2>Categorical Variables and One Hot Encoding</h2>

```python


"""
 Q) how should we handle text data in numeric model?

     - interger encoding or label encoding
     - where we convert name of town to a specific integer number

Categorical variables

    - Nominal
    - Ordinal


Nominal

 ex: 

    1) male, female
    2) green, red, yellow
    3) monroe township, west windsor, robinsville - cities name

    - when we convert above names to integer, there is no numerical order is there
      so, model will confuse
    
Ordinal 

  ex: 

    1) satisfied, neutral, dissatisfied  => satisfied > neutral > dissatisfied
    2) graduate, masters, phd  => graduate < masters < phd
    3) high, mdeium, low => high > mdeium > low

    - when we convert above names to integer, there is numerical order between 
      them. so model will understand
    
"""
```

```python
"""
 - for this nominal dataset we are dealing with nominal categorical variables and
   simple integer encoding is not going to work

 - so we are going use technique called one hot encoding to solve our problem,
   the variables we are creating for encoding 0, 1 also called dummy variables
 
"""
```

<h2 style='color:purple'>1. Using pandas to create dummy variables</h2>

```python
import pandas as pd


df = pd.read_csv("homeprices_one_hot_encoding.csv")
df
```

```python
dummies = pd.get_dummies(df.town, dtype=int)
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')  # axis - merge using columns
merged
```

<h3 style='color:purple'>Dummy Variable Trap</h3>

```python
"""
 - drop town column(text column), we dont need anymore because we have dummy variable column(encoded column)
 - drop one these dummy variable column


Explanation for droping dummy variable column:

  Dummy variable trap:

  - whenever one variable can be derived from rest of the variables, these variables are said to be
    multi-co-linear, and whenever we have multi-co-linearity in our dataset it creates the probelm of 
    dummy variable trap that can mess up our ML model

  - so the rule is to drop one of the dummy variabe columns so if we have lets say 5 dummy variable then drop one 
    then we will 4 columns. here we have 3 dummy variables columns(monroe township, robinsville, west windsor)
    so we are just going to remove one of them, we can remove any column out of these three.
 
"""
```

```python
final = merged.drop(['town', 'west windsor'], axis='columns')
final
```


**Note:  when we are using sklearn linear regression model it will work even if you dont drop it becase linear regreesion
   model is aware about dummy variable trap and it will drop it automatically. but in general its a good practice that 
   we drop it on our own**


```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
input_x = final.drop('price', axis='columns')
input_x
```

```python
output_y = final.price
output_y
```

```python
model.fit(input_x, output_y)  # training the data
```

```python
model.predict([[2800, 0, 1]]) # predict area price in robinsville
```

```python
model.predict([[3400, 0, 0]]) # predict area price in west windsor
```

```python
# check model score

model.score(input_x, output_y)
```

<h2 style='color:purple'>2. Using sklearn OneHotEncoder</h2>

First step is to use label encoder to convert town names into numbers

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
dfle = df

dfle_town = le.fit_transform(dfle.town)  # 0 - monroe township, 1- robinsville , 2 - west windsor
dfle_town
```

```python
dfle.town = dfle_town
dfle
```

```python
input_x = dfle[['town', 'area']].values  # training dataset
input_x
```

```python
output_y = dfle.price.values
output_y
```

Now use one hot encoder to create dummy variables for each of the town

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ohe = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(sparse_output=False), [0])   # it take 0th column as categorical variable - town column
    ],
    remainder='passthrough'  # Keeps the other columns
) 
```

```python
input_x = ohe.fit_transform(input_x)  # [0] -> [1.0e+00(monroe township), 0.0e+00(robinsville), 0.0e+00(west windsor)], [1]-> [2.6e+03(area)]		
input_x
```

```python
# drop one dummy variable column

input_x = input_x[:,1:]  # droped first dummy variable column(monroe township)
input_x
```

```python
model.fit(input_x, output_y) 
```

```python
model.predict([[1,0, 2800]])  # predict area price in robinsville 
```

```python
model.predict([[0,1, 3400]])  # predict area price in west windsor 
```
